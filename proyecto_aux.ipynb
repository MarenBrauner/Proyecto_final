```python
import geopandas as gpd

df_grafico2['geocode'] = '2024' + df_grafico2['geocode'].str[4:]

gdf_tenerife_2023 = gpd.read_file('data/secciones_20240101_tenerife.json')
gdf_tenerife_2023['geocode'].dtype

gdf_final_map = pd.merge(gdf_tenerife_2023, df_grafico2, on='geocode', how='left')

gdf_final_map.dtypes

nulos_renta_raw = gdf_final_map['OBS_VALUE'].isna().sum()
total_secciones_raw = len(gdf_final_map)
    
    # 2. CASTEO ESTRICTO A TIPOS NATIVOS DE PYTHON (Para contentar a Dagster)
nulos_renta = int(nulos_renta_raw)
total_secciones = int(total_secciones_raw)
    
    # 3. Cálculo de la tasa de cobertura
if total_secciones == 0:
    tasa_cobertura = 0.0
else:
    tasa_cobertura = (total_secciones - nulos_renta) / total_secciones

tasa_cobertura

passed = tasa_cobertura >= 0.90

passed
```

```python
df_municipio_sector = (
    df_actividad_clean.groupby(['municipio', 'Actividad económica'], as_index=False)
    ['num_casos'].sum()
)

# 3. Encontrar la fila con el máximo número de casos para cada municipio
# idxmax() identifica el índice del valor más alto dentro de cada grupo de municipios
idx_maximo = df_municipio_sector.groupby('municipio')['num_casos'].idxmax()
df_sector_mayoritario = df_municipio_sector.loc[idx_maximo].copy()

# 4. Ordenar el resultado por volumen para facilitar el análisis visual
df_sector_mayoritario = df_sector_mayoritario.sort_values(by='num_casos', ascending=False)

# Mostrar el resultado
print(df_sector_mayoritario.to_string(index=False))
print(df_sector_mayoritario['Actividad económica'].unique())
```
